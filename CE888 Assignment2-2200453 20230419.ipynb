```python
score_dir='C:/Users/Admin/Documents/1.MSc_Data science/Data science&decision making/Assignment/Questionaire/'

score_pre=pd.read_csv(score_dir+'Questionnaire_datasetIA.csv',encoding="cp1252")
score_post=pd.read_csv(score_dir+'Questionnaire_datasetIB.csv',encoding="cp1252")
```

```python
score=pd.DataFrame({'Participant name':df_proceed['Participant name'].unique(),
            'Score_original':score_post['Total Score original'],
            'Score_extended':score_post['Total Score extended']})
```

```python
data= pd.merge(df_proceed, score, on='Participant name')
```

```python
data
```

```python
X = data.drop(['Participant name','Project name','Recording name', 'Score_original', 'Score_extended'], axis=1)
y = data['Score_extended']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor


# Define the models
models = [
    ('LR', LinearRegression()),
    ('DT', DecisionTreeRegressor()),
    ('RF', RandomForestRegressor(n_estimators=100)),
    ('SVM', SVR(kernel='linear')),
    ('KNN', KNeighborsRegressor(n_neighbors=2))
]

# Evaluate each model using cross-validation
results = []
names = []
for name, model in models:
    cv = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
    rmse = np.sqrt(-cv)
    results.append(rmse)
    names.append(name)
    print(f"{name}: Mean RMSE = {round(rmse.mean(),3)}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR


# Define the models
models = [
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('Random Forest', RandomForestRegressor(n_estimators=100)),
    ('SVM', SVR(kernel='linear'))
]

# Evaluate each model using RMSE
results = []
names = []
for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    results.append(rmse)
    names.append(name)
    print(f"{name}: RMSE = {round(rmse, 3)}")

# Choose the best model
best_model_idx = results.index(min(results))
best_model_name = names[best_model_idx]
best_model = models[best_model_idx][1]
print(f"The best model is {best_model_name} with RMSE = {round(min(results), 3)}")
```
